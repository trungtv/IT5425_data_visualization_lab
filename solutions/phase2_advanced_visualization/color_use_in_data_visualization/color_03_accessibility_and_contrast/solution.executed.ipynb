# Color 03 - Accessibility and contrast checks (Solution)

```python
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError("Cannot locate data/gapminder.csv")

root = resolve_repo_root()
sns.set_theme(style="whitegrid")
df = pd.read_csv(root / "data" / "gapminder.csv")
d2007 = df[df["year"] == 2007].copy()
```

## 1) Dung palette than thien color-blind

```python
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.scatterplot(data=d2007, x="gdpPercap", y="lifeExp", hue="continent", palette="deep", ax=axes[0])
axes[0].set_xscale("log"); axes[0].set_title("Default deep")
sns.scatterplot(data=d2007, x="gdpPercap", y="lifeExp", hue="continent", palette="colorblind", ax=axes[1])
axes[1].set_xscale("log"); axes[1].set_title("Seaborn colorblind palette")
plt.tight_layout(); plt.show()
```

## 2) Contrast ratio helper (WCAG) cho text tren nen mau

```python
def srgb_to_linear(c):
    c = c / 255.0
    return c/12.92 if c <= 0.04045 else ((c+0.055)/1.055)**2.4

def luminance(rgb):
    r,g,b = [srgb_to_linear(v) for v in rgb]
    return 0.2126*r + 0.7152*g + 0.0722*b

def contrast_ratio(rgb1, rgb2):
    L1, L2 = luminance(rgb1), luminance(rgb2)
    Lmax, Lmin = max(L1, L2), min(L1, L2)
    return (Lmax + 0.05) / (Lmin + 0.05)

white = (255,255,255)
blue = (33,113,181)
light_blue = (198,219,239)
print("white vs blue:", round(contrast_ratio(white, blue),2))
print("white vs light_blue:", round(contrast_ratio(white, light_blue),2))
```

## 3) Redundant encoding: mau + shape

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=d2007, x="gdpPercap", y="lifeExp", hue="continent", style="continent", palette="colorblind")
plt.xscale("log")
plt.title("Hue + style for robust accessibility")
plt.tight_layout(); plt.show()
```
