# Color 04 - Color in context (Solution)

```
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError("Cannot locate data/gapminder.csv")

root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
d2007 = df[df["year"] == 2007].copy()
```

## 1) Choropleth map: sequential cho magnitude

```
fig = px.choropleth(
    d2007,
    locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    color_continuous_scale="Viridis",
    title="Life expectancy (2007) - sequential scale"
)
fig.show()
```

## 2) Heatmap trend theo continent-year: sequential vs diverging

```
p = df.groupby(["continent","year"], as_index=False)["lifeExp"].mean().pivot(index="continent", columns="year", values="lifeExp")
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.heatmap(p, cmap="mako", ax=axes[0]); axes[0].set_title("Absolute values -> sequential")
sns.heatmap(p - p.mean().mean(), cmap="RdBu_r", center=0, ax=axes[1]); axes[1].set_title("Deviation around center -> diverging")
plt.tight_layout(); plt.show()
```

## 3) Scatter highlight: neutral gray + highlight color

```
focus = d2007["country"].isin(["Vietnam", "United States", "China", "India"])
plt.figure(figsize=(9,5))
plt.scatter(d2007.loc[~focus, "gdpPercap"], d2007.loc[~focus, "lifeExp"], color="lightgray", alpha=0.7, label="others")
plt.scatter(d2007.loc[focus, "gdpPercap"], d2007.loc[focus, "lifeExp"], color="#d62728", alpha=0.95, label="focus")
plt.xscale("log"); plt.legend(); plt.title("Neutral + highlight strategy")
plt.tight_layout(); plt.show()
```
