# Perceptual Diagnostics and Chart Critique (Solution)

Detect misleading encodings: truncated axis, dual-axis confusion, overplotting.

```
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

sns.set_theme(style="whitegrid")

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError("Cannot locate data/gapminder.csv")

root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
d2007 = df[df["year"] == 2007].copy()
```

## 1) Truncated axis diagnostic

```
agg = d2007.groupby('continent',as_index=False)['lifeExp'].mean().sort_values('lifeExp')
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.barplot(data=agg, x='continent', y='lifeExp', ax=axes[0])
axes[0].set_ylim(0, 90)
axes[0].set_title('Full axis')
sns.barplot(data=agg, x='continent', y='lifeExp', ax=axes[1])
axes[1].set_ylim(65, 80)
axes[1].set_title('Truncated axis (can mislead)')
for ax in axes: ax.tick_params(axis='x', rotation=20)
plt.tight_layout(); plt.show()
```

## 2) Overplotting vs transparency

```
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.scatterplot(data=d2007, x='gdpPercap', y='lifeExp', ax=axes[0])
axes[0].set_xscale('log'); axes[0].set_title('Opaque points')
sns.scatterplot(data=d2007, x='gdpPercap', y='lifeExp', alpha=0.35, ax=axes[1])
axes[1].set_xscale('log'); axes[1].set_title('Transparency improves perception')
plt.tight_layout(); plt.show()
```
