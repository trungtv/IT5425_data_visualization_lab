# High-dimensional Projection Charts (Solution)

Explore high-dimensional data with parallel coordinates and PCA-like projection.

```python
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

sns.set_theme(style="whitegrid")

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError("Cannot locate data/gapminder.csv")

root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
d2007 = df[df["year"] == 2007].copy()
```

## 1) Parallel coordinates

```python
pc = d2007[['country','continent','lifeExp','gdpPercap','pop']].copy()
pc['log_gdp'] = np.log10(pc['gdpPercap'])
pc['log_pop'] = np.log10(pc['pop'])
fig = px.parallel_coordinates(pc, dimensions=['lifeExp','log_gdp','log_pop'], color='lifeExp')
fig.show()
```

## 2) PCA-like 2D projection via SVD

```python
X = d2007[['lifeExp','gdpPercap','pop']].copy()
X['gdpPercap'] = np.log10(X['gdpPercap'])
X['pop'] = np.log10(X['pop'])
X = (X - X.mean()) / X.std()
U, S, Vt = np.linalg.svd(X.to_numpy(), full_matrices=False)
Z = U[:, :2] * S[:2]
proj = d2007[['country','continent']].copy()
proj['pc1'] = Z[:,0]
proj['pc2'] = Z[:,1]
plt.figure(figsize=(7,5))
sns.scatterplot(data=proj, x='pc1', y='pc2', hue='continent', alpha=0.8)
plt.title('PCA-like projection (SVD)')
plt.tight_layout(); plt.show()
```
