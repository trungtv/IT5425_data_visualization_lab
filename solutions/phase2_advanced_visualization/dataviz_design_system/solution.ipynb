# Dataviz Design System and Styling (Solution)

Build a consistent visual language: typography, spacing, palette, annotation hierarchy.

```
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

sns.set_theme(style="whitegrid")

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError("Cannot locate data/gapminder.csv")

root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
d2007 = df[df["year"] == 2007].copy()
```

## 1) Before/after style system

```
agg = df.groupby(['year','continent'],as_index=False)['lifeExp'].mean()
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.lineplot(data=agg, x='year', y='lifeExp', hue='continent', ax=axes[0])
axes[0].set_title('Default style')

sns.set_theme(style='white', context='talk', palette='colorblind')
sns.lineplot(data=agg, x='year', y='lifeExp', hue='continent', ax=axes[1])
axes[1].set_title('Design-system style')
for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
plt.tight_layout(); plt.show()
```

## 2) Annotation hierarchy

```
asia = agg[agg['continent']=='Asia']
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(asia['year'], asia['lifeExp'], color='#1f77b4', lw=2)
max_row = asia.loc[asia['lifeExp'].idxmax()]
ax.scatter([max_row['year']], [max_row['lifeExp']], color='#d62728', zorder=3)
ax.annotate('Peak', (max_row['year'], max_row['lifeExp']), xytext=(max_row['year']-8, max_row['lifeExp']+1.2), arrowprops=dict(arrowstyle='->'))
ax.set_title('Annotation hierarchy example')
plt.tight_layout(); plt.show()
```
