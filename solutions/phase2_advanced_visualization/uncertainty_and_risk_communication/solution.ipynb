# Uncertainty and Risk Communication (Solution)

Visualize uncertainty with error bars, confidence bands, and scenario fans.

```
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

sns.set_theme(style="whitegrid")

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError("Cannot locate data/gapminder.csv")

root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
d2007 = df[df["year"] == 2007].copy()
```

## 1) Error bar chart theo nhom

```
agg = d2007.groupby("continent", as_index=False)["lifeExp"].agg(mean="mean", std="std", count="count")
agg["se"] = agg["std"] / np.sqrt(agg["count"])
agg["ci95"] = 1.96 * agg["se"]
plt.figure(figsize=(8,4))
plt.errorbar(agg["continent"], agg["mean"], yerr=agg["ci95"], fmt='o', capsize=4)
plt.title("Mean lifeExp with 95% CI")
plt.tight_layout(); plt.show()
```

## 2) Confidence band theo thoi gian

```
trend = df.groupby(["year","continent"], as_index=False)["lifeExp"].mean()
cont = "Asia"
x = trend[trend["continent"]==cont]["year"].to_numpy()
y = trend[trend["continent"]==cont]["lifeExp"].to_numpy()
# toy uncertainty band
band = np.linspace(0.8, 1.3, len(x))
plt.figure(figsize=(8,4))
plt.plot(x,y,label=cont)
plt.fill_between(x, y-band, y+band, alpha=0.2, label="uncertainty band")
plt.legend(); plt.title("Trend + uncertainty band")
plt.tight_layout(); plt.show()
```

## 3) Scenario fan chart (toy)

```
years = np.arange(2008, 2021)
base = np.linspace(df[df["year"]==2007]["lifeExp"].mean(), 75, len(years))
low = base - np.linspace(1.5, 3.0, len(years))
high = base + np.linspace(1.0, 2.2, len(years))
plt.figure(figsize=(8,4))
plt.plot(years, base, color='black', label='expected')
plt.fill_between(years, low, high, color='#4C78A8', alpha=0.25, label='range')
plt.legend(); plt.title("Scenario fan (illustrative)")
plt.tight_layout(); plt.show()
```
