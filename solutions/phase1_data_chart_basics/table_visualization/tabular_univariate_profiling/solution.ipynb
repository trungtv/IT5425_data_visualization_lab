# Phase 1 — Tabular data: univariate profiling (Solution)

**Mục tiêu:** khi DataFrame có nhiều cột (số/rời rạc), trả lời nhanh *mỗi biến đang phân bố ra sao* bằng chart phù hợp Stevens/EDA.

Dữ liệu: `data/gapminder.csv`.


```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pathlib import Path

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / "data" / "gapminder.csv").exists():
            return p
    raise FileNotFoundError(
        "Cannot locate data/gapminder.csv from current working directory"
    )

root = resolve_repo_root()


sns.set_theme(style="whitegrid")
df = pd.read_csv(root / "data" / "gapminder.csv")
df.head()
```

## 1) Numeric: histogram + KDE + ECDF

Với biến liên tục (`lifeExp`, `gdpPercap`), dùng nhiều góc nhìn để tránh “một chart một câu chuyện”.


```
d2007 = df[df["year"] == 2007].copy()

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
sns.histplot(d2007["lifeExp"], kde=True, ax=axes[0])
axes[0].set_title("Histogram + KDE (lifeExp)")

sns.kdeplot(d2007["lifeExp"], ax=axes[1], fill=True)
axes[1].set_title("KDE only")

sns.ecdfplot(d2007["lifeExp"], ax=axes[2])
axes[2].set_title("ECDF (percentile map)")
plt.tight_layout()
plt.show()
```

## 2) Categorical: tần suất (`countplot`)

Biến nominal/ordinal trong bảng thường là `continent`, `country` — ưu tiên *count/frequency* trước khi vẽ phức tạp.


```
fig, ax = plt.subplots(figsize=(7, 4))
sns.countplot(data=d2007, x="continent", ax=ax, order=d2007["continent"].value_counts().index)
ax.set_title("Số quốc gia theo continent (2007)")
ax.tick_params(axis="x", rotation=20)
plt.tight_layout()
plt.show()
```

## 3) So sánh phân phối theo nhóm: box + violin

Giúp thấy median/IQR và hình dạng phân phối khi có cột nhóm trong cùng bảng.


```
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.boxplot(data=d2007, x="continent", y="lifeExp", ax=axes[0])
axes[0].tick_params(axis="x", rotation=20)

sns.violinplot(data=d2007, x="continent", y="lifeExp", ax=axes[1], inner="quartile")
axes[1].tick_params(axis="x", rotation=20)
plt.tight_layout()
plt.show()
```

## 4) Missingness (nếu có)

`gapminder` sạch; pattern này áp dụng cho bảng thực tế.

**Việc của bạn khi làm lab:** nhận xét scale (ratio vs interval) trước khi chọn biểu đồ.


```
missing_rate = df.isna().mean().sort_values(ascending=False)
missing_rate.head(10)
```
