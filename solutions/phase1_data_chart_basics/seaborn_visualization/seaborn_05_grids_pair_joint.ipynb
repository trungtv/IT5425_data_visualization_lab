# Seaborn 05 - Multi-Plot Grids (Solution)

Noi dung duoc tham khao tu Seaborn docs 0.13.x (function overview, relational, categorical, grids, objects).

**Trong tam:** FacetGrid, PairGrid/pairplot, jointplot cho exploratory analysis nhanh.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
sns.set_theme(style='ticks')
from pathlib import Path
def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')
root = resolve_repo_root()
df = pd.read_csv(root / 'data' / 'gapminder.csv')
d2007 = df[df['year']==2007].copy()
```

```
g = sns.FacetGrid(d2007, col='continent', col_wrap=3, height=3.0, sharex=False, sharey=False)
g.map_dataframe(sns.histplot, x='lifeExp', bins=12)
g.fig.suptitle('FacetGrid: lifeExp per continent', y=1.03)
plt.show()
```

```
subset = d2007[['lifeExp','gdpPercap','pop','continent']].copy()
subset['log_gdp'] = subset['gdpPercap'].map(lambda x: __import__('math').log10(x+1))
subset['log_pop'] = subset['pop'].map(lambda x: __import__('math').log10(x+1))
p = sns.pairplot(subset[['lifeExp','log_gdp','log_pop','continent']], hue='continent', corner=True, plot_kws={'alpha':0.7, 's':30})
p.fig.suptitle('pairplot (corner)', y=1.02)
plt.show()
```

```
j = sns.jointplot(data=d2007, x='gdpPercap', y='lifeExp', kind='hex', height=6)
j.ax_joint.set_xscale('log')
j.fig.suptitle('jointplot (hex)', y=1.02)
plt.show()
```
