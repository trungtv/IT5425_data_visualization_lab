# Seaborn Visualization Basics (Solution)

Muc tieu: Biểu đồ Seaborn thực chiến: distribution, comparison, relationship, faceting.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
sns.set_theme(style="whitegrid")
from pathlib import Path
def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')
root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
df_2007 = df[df["year"] == 2007].copy()
df_2007.head()
```

```
plt.figure(figsize=(8,4))
sns.histplot(data=df_2007, x="lifeExp", bins=20, kde=True)
plt.title("Life Expectancy Distribution (2007)")
plt.tight_layout(); plt.show()
```

```
top = (df_2007.groupby("continent", as_index=False)["gdpPercap"].mean())
plt.figure(figsize=(7,4))
sns.barplot(data=top, x="continent", y="gdpPercap")
plt.title("Average GDP per Capita by Continent (2007)")
plt.tight_layout(); plt.show()
```

```
g = sns.relplot(
    data=df_2007,
    x="gdpPercap", y="lifeExp",
    hue="continent", size="pop",
    kind="scatter", height=5, aspect=1.3
)
g.set(xscale="log")
plt.title("GDP vs Life Expectancy")
plt.show()
```
