# Seaborn 04 - Categorical Comparisons (Solution)

Noi dung duoc tham khao tu Seaborn docs 0.13.x (function overview, relational, categorical, grids, objects).

**Trong tam:** barplot/pointplot/boxplot/violinplot/swarmplot va uoc luong uncertainty.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
sns.set_theme(style='whitegrid')
from pathlib import Path
def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')
root = resolve_repo_root()
df = pd.read_csv(root / 'data' / 'gapminder.csv')
d2007 = df[df['year']==2007].copy()
```

```
plt.figure(figsize=(8,4))
sns.barplot(data=d2007, x='continent', y='lifeExp', estimator='mean', errorbar=('ci',95))
plt.title('barplot with CI')
plt.tight_layout(); plt.show()
```

```
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.boxplot(data=d2007, x='continent', y='lifeExp', ax=axes[0])
axes[0].set_title('boxplot')
sns.violinplot(data=d2007, x='continent', y='lifeExp', inner='quart', ax=axes[1])
axes[1].set_title('violinplot')
plt.tight_layout(); plt.show()
```

```
plt.figure(figsize=(8,4))
sns.pointplot(data=d2007, x='continent', y='gdpPercap', estimator='median', errorbar=('pi',50))
plt.title('pointplot: median + interval')
plt.tight_layout(); plt.show()
```
