# Seaborn 02 - Distribution Analysis (Solution)

Noi dung duoc tham khao tu Seaborn docs 0.13.x (function overview, relational, categorical, grids, objects).

**Trong tam:** histplot, kdeplot, ecdfplot, displot va so sanh phan phoi theo nhom.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
sns.set_theme(style='whitegrid')
from pathlib import Path
def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')
root = resolve_repo_root()
df = pd.read_csv(root / 'data' / 'gapminder.csv')
d2007 = df[df['year']==2007].copy()
```

```
plt.figure(figsize=(8,4))
sns.histplot(data=d2007, x='lifeExp', hue='continent', bins=20, element='step', stat='density', common_norm=False)
plt.title('histplot: lifeExp distribution by continent')
plt.tight_layout(); plt.show()
```

```
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.kdeplot(data=d2007, x='lifeExp', hue='continent', ax=axes[0])
axes[0].set_title('kdeplot')
sns.ecdfplot(data=d2007, x='lifeExp', hue='continent', ax=axes[1])
axes[1].set_title('ecdfplot')
plt.tight_layout(); plt.show()
```

```
g = sns.displot(data=d2007, x='gdpPercap', col='continent', col_wrap=3, facet_kws={'sharex':False, 'sharey':False}, bins=16)
g.fig.suptitle('displot with faceting', y=1.03)
plt.show()
```
