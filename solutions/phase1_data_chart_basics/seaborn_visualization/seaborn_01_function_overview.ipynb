# Seaborn 01 - Function Overview (Solution)

Noi dung duoc tham khao tu Seaborn docs 0.13.x (function overview, relational, categorical, grids, objects).

**Trong tam:** Phan biet figure-level va axes-level; relplot/displot/catplot vs scatterplot/histplot/barplot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
sns.set_theme(style='whitegrid')

def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')

root = resolve_repo_root()
df = pd.read_csv(root / 'data' / 'gapminder.csv')
d2007 = df[df['year']==2007].copy()
d2007.head()
```

```python
# Axes-level
fig, ax = plt.subplots(figsize=(7,4))
sns.scatterplot(data=d2007, x='gdpPercap', y='lifeExp', hue='continent', ax=ax)
ax.set_xscale('log')
ax.set_title('Axes-level: scatterplot')
plt.tight_layout(); plt.show()
```

```python
# Figure-level
g = sns.relplot(data=d2007, x='gdpPercap', y='lifeExp', hue='continent', size='pop', kind='scatter', height=4.5, aspect=1.4)
g.set(xscale='log')
g.fig.suptitle('Figure-level: relplot', y=1.02)
plt.show()
```
