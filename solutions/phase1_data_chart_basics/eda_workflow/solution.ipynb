Muc tieu: Quy trình EDA hoàn chỉnh: quality checks -> univariate -> bivariate -> insight summary.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
sns.set_theme(style="whitegrid")
from pathlib import Path
def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')
root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
df.shape, df.isna().sum().sum()
```

```
eda = df.copy()
eda["log_gdp"] = (eda["gdpPercap"] + 1).map(lambda x: __import__("math").log10(x))
eda[["lifeExp", "gdpPercap", "pop"]].describe().T
```

```
fig, axes = plt.subplots(1, 2, figsize=(12,4))
sns.boxplot(data=eda, x="continent", y="lifeExp", ax=axes[0])
axes[0].set_title("LifeExp by Continent")
sns.scatterplot(data=eda[eda["year"]==2007], x="log_gdp", y="lifeExp", hue="continent", ax=axes[1])
axes[1].set_title("2007: log10(GDP) vs LifeExp")
plt.tight_layout(); plt.show()
```

```
insights = pd.DataFrame({
    "insight": [
        "Life expectancy differs notably by continent.",
        "Higher GDP per capita is associated with higher life expectancy.",
        "Population scale varies widely, so log transform/size encoding is useful."
    ]
})
insights
```
