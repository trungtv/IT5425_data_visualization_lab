# Pandas Core Skills (Solution)

Muc tieu: DataFrame thao tac cot lõi: load, filter, groupby, merge, missing values.

```
import pandas as pd
from pathlib import Path
from pathlib import Path
def resolve_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'data' / 'gapminder.csv').exists():
            return p
    raise FileNotFoundError('Cannot locate data/gapminder.csv from current working directory')
root = resolve_repo_root()
df = pd.read_csv(root / "data" / "gapminder.csv")
df.head()
```

```
cols = ["country", "continent", "year", "lifeExp", "gdpPercap", "pop"]
work = df[cols].copy()
work = work[work["year"].between(1992, 2007)]
work.head()
```

```
summary = (
    work.groupby(["continent", "year"], as_index=False)
        .agg(avg_lifeExp=("lifeExp", "mean"), avg_gdp=("gdpPercap", "mean"), total_pop=("pop", "sum"))
)
summary.tail()
```

```
latest = summary[summary["year"] == summary["year"].max()]
latest.sort_values("avg_lifeExp", ascending=False)
```
