# Lab 05 — Figure design principles (Solution)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Before/after cho overlapping points

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharex=True, sharey=True)
axes[0].scatter(tips['total_bill'], tips['tip'])
axes[0].set_title('Before: raw scatter')
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='time',
    alpha=0.7,
    ax=axes[1],
)
axes[1].set_title('After: color + alpha')
plt.tight_layout()
plt.show()
```

Small multiples

```python
g = sns.FacetGrid(tips, col='sex', row='smoker', height=3)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', alpha=0.7)
g.fig.suptitle('Small multiples by sex/smoker', y=1.02)
plt.show()
```

## Reflection
- Data-ink ratio tốt hơn khi loại bỏ yếu tố trang trí không cần thiết.
- Small multiples giúp so sánh nhóm rõ hơn overlay quá nhiều màu.


## Extended analysis — chart quality checklist
Checklist tự động hoá một phần cho figure design.

```python
checklist = {
    'has_title': True,
    'axis_labeled': True,
    'legend_meaningful': True,
    'overplotting_mitigated': True,
    'no_3d_chartjunk': True,
}
pd.Series(checklist, name='pass')
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
summary = tips.groupby('day', as_index=False).agg(avg_tip_pct=('tip_pct', 'mean'))
summary
```

```python
plt.figure(figsize=(6, 4))
plt.bar(summary['day'], summary['avg_tip_pct'])
plt.title('Average tip ratio by day')
plt.ylabel('tip / total_bill')
plt.tight_layout(); plt.show()
```
